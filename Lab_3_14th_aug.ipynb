```python
# Matrix Addition and Subtraction

import numpy as np

# Create two 3x3 NumPy arrays with random integers (0–9)
A = np.random.randint(0, 10, (3, 3))
B = np.random.randint(0, 10, (3, 3))

# Perform element-wise addition and subtraction using NumPy functions
addition_result = np.add(A, B)
subtraction_result = np.subtract(A, B)

# Display original matrices and results
print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
print("\nAddition (A + B):\n", addition_result)
print("\nSubtraction (A - B):\n", subtraction_result)

# Verify one element manually
row, col = 1, 2  # pick an element for verification
manual_add = A[row, col] + B[row, col]
manual_sub = A[row, col] - B[row, col]

print(f"\nManual check for element at position ({row},{col}):")
print(f"Addition: {A[row, col]} + {B[row, col]} = {manual_add}")
print(f"Subtraction: {A[row, col]} - {B[row, col]} = {manual_sub}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create an "image" matrix A (3 pixels, 2 color channels → like a small 2-channel image)
A = np.random.randint(0, 256, (3, 2), dtype=np.uint8)  # pixel values 0–255

# Matrix B: Transformation matrix (2 color channels → 3 color channels, like RGB)
B = np.random.randint(0, 10, (2, 3))

# Step 2: Print matrices
print("Matrix A (3x2 - Image Data):\n", A)
print("\nMatrix B (2x3 - Transformation Filter):\n", B)

# Step 3: Perform matrix multiplication
result = np.dot(A, B)  # (3x2) · (2x3) → (3x3)
print("\nMatrix Multiplication Result (3x3 - Transformed Image):\n", result)

# Step 4: Verify one element manually
row, col = 1, 2
manual_calc = A[row, 0] * B[0, col] + A[row, 1] * B[1, col]
print(f"\nManual verification for element ({row},{col}):")
print(f"{A[row,0]}*{B[0,col]} + {A[row,1]}*{B[1,col]} = {manual_calc}")

# Step 5: Simulate visual transformation
plt.figure(figsize=(8, 3))

plt.subplot(1, 3, 1)
plt.imshow(A, cmap='plasma', interpolation='nearest')
plt.title("Matrix A (Input Image)")
plt.colorbar()

plt.subplot(1, 3, 2)
plt.imshow(B, cmap='cool', interpolation='nearest')
plt.title("Matrix B (Filter)")
plt.colorbar()

plt.subplot(1, 3, 3)
plt.imshow(result, cmap='viridis', interpolation='nearest')
plt.title("Result (Transformed)")
plt.colorbar()

plt.tight_layout()
plt.show()
```
