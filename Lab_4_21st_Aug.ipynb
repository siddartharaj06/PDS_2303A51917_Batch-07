Student Marks Analysis
Objective:
Clean and analyze student performance data by filling missing marks
and filtering high achievers.
Requirements:
• Python
• pandas library
• student_scores.csv file containing columns: Student_ID, Name,
Math, Science, English
Deliverables:
• Read the CSV file
• Fill missing subject marks with column-wise average
• Filter and display students who scored more than 75 in all
subjects


```python
import pandas as pd

df = pd.read_csv("/content/student-scores.csv")
```

```python
subjects = ["math_score", "history_score", "physics_score",
            "chemistry_score", "biology_score", "english_score",
            "geography_score"]

df[subjects] = df[subjects].fillna(df[subjects].mean())


high_achievers = df[df[subjects].gt(75).all(axis=1)]
```

```python
print("High Achievers (All subjects > 75):")
print(high_achievers)

high_achievers.to_csv("high_achievers.csv", index=False)
files.download("high_achievers.csv")
```

Hospital Patient Records
Objective:
Clean hospital patient data and extract records of elderly female
patients diagnosed with diabetes.
Requirements:
• Python
• pandas library
• patients.csv with columns: Patient_ID, Name, Age, Gender,
Diagnosis, Admission_Date
Deliverables:
• Load the dataset
• Replace missing Age values with median
• Filter female patients aged above 50 diagnosed with Diabetes



```python
import pandas as pd

# Parse Admission_Date as a date; coerce Age to numeric
df = pd.read_csv("/content/patients.csv", parse_dates=["Admission_Date"])
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

print(df.head())
```

```python
median_age = df["Age"].median(skipna=True)
df["Age"] = df["Age"].fillna(median_age)

print("Median age used for imputation:", median_age)
df.isna().sum()  # quick null check
```

```python
filtered = df[
    (df["Gender"].str.lower() == "female") &
    (df["Age"] > 50) &
    (df["Diagnosis"].str.contains("Diabetes", case=False, na=False))
].copy()

print("Total records after filtering:", len(filtered))
filtered
```

```python
from google.colab import files

filtered.to_csv("elderly_female_diabetes.csv", index=False)
files.download("elderly_female_diabetes.csv")
```
